```python
import numpy as np
import matplotlib.pyplot as plt
from qclab.models.holstein_lattice import HolsteinLatticeModel
from qclab.recipes import MeanFieldDynamicsRecipe
from qclab.drivers.ray_driver import dynamics_parallel_ray
from qclab.drivers.serial_driver import dynamics_serial
```

```python
import numpy as np
import matplotlib.pyplot as plt   
from qclab.models.holstein_lattice import HolsteinLatticeModel

import qclab.auxiliary as auxiliary
input_params = dict(num_states=10, temp = 1, j=1, g=5, w=0.2, m=1, open_boundaries=False)
model = HolsteinLatticeModel(input_params = input_params)



model.wf_db = np.zeros((model.num_states),dtype=complex)
# initialize a quantum state in the upper state
model.wf_db[0] = 1

recipe = MeanFieldDynamicsRecipe()
recipe.params.batch_size = 1
recipe.params.tmax=20
recipe.params.dt_output=0.0001
recipe.params.dt=0.0001
recipe.params.num_trajs=1
recipe.params.batch_size=1

def get_evals(state, model, params):
    state.evals, _ = np.linalg.eigh(state.h_quantum)
    return state, model, params

recipe.output.append(get_evals)
recipe.output_names.append('evals')

data_mf = dynamics_serial(recipe=recipe, model = model,ncpus=1, data = auxiliary.Data())
```

```python
evals = data_mf.data_dic['evals'][:,0,0,:]
print(np.min(np.abs(evals[:,1:] - evals[:,:-1])))
plt.plot(np.abs(evals[:,1:] - evals[:,:-1]))
plt.show()
```
